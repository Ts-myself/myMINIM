```python
import os
import numpy as np
import PIL
from PIL import Image
import matplotlib.pyplot as plt


number = 0
# 67 is the best

def shift(id, number):
    return (id+number)%400


image_dir = '/mnt/f/dataset/OCTA_500_BOTH_MODAL/OCT/10239'

image_path = os.listdir(image_dir)

image_path.sort(key= lambda x: int(x.split('/')[-1].split('.')[0]))

print(image_path[0])
shape = Image.open(os.path.join(image_dir, image_path[0])).size

image_2D = np.zeros((400, 400))

plt.figure(figsize=(20, 20))


count = 0
for i in image_path:
    print(count)
    image = Image.open(os.path.join(image_dir, i))
    image = np.array(image)
    image_2D[shift(count, number)] = np.max(image, axis=0)
    count += 1


```

```python
# 添加图像到大画板
# 上下翻转图片
image_2D = np.flipud(image_2D)


plt.imshow(image_2D, cmap='gray')
plt.title(f"shift {number}")
plt.show()
```
